# Exercise 9

## Mashable news stories analysis

Predicting if a news story is going to be popular


```python
import pandas as pd
import numpy as np
url = 'https://raw.githubusercontent.com/albahnsen/PracticalMachineLearningClass/master/datasets/mashable.csv'
df = pd.read_csv(url, index_col=0)
df.head()
```

```python
df.shape
```

```python
X = df.drop(['url', 'Popular'], axis=1)
y = df['Popular']
```

```python
y.mean()
```

```python
# train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
y_train
```

# Exercise 9.1

Estimate a Decision Tree Classifier and a Logistic Regresion

Evaluate using the following metrics:
* Accuracy
* F1-Score

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
import warnings
warnings.filterwarnings('ignore')

models = {'lr': LogisticRegression(),
          'dt': DecisionTreeRegressor()}
```

```python
for model in models.keys():
    models[model].fit(X_train, y_train)
```

```python
# predict test for each model
y_pred = pd.DataFrame(index=X_test.index, columns=models.keys())
for model in models.keys():
    y_pred[model] = models[model].predict(X_test)
```

```python
# Evaluate each model
from sklearn.metrics import mean_squared_error

for model in models.keys():
    print(model,np.sqrt(mean_squared_error(y_pred[model], y_test)))
```

```python
y_pred.mean(axis=1)
```

```python
from sklearn import metrics
print("\033[1;38mF1 score logistic regression:",metrics.f1_score(y_pred['lr'], y_test))
print("\033[1;38mAccuracy logistic regression:",metrics.accuracy_score(y_pred['lr'], y_test))
```

```python
print("\033[1;38mF1 score decision tree regression:",metrics.f1_score(y_pred['dt'], y_test))
print("\033[1;38mAccuracy linear regression:",metrics.accuracy_score(y_pred['dt'], y_test))
```

Para los dos modelos, regresion logística y árboles de decisión, se encontraron valores cercanos para las métricas de F1-score y accuracy. Lo cual nos da confianza para el análisis de los resultados obtenido. Ahora, si comparamos los valores entre los dos modelos, la regresión logística obtuvo valores cercanos a 0.60 mientras que los árboles de decisión obtuvieron un valor de 0.57. De tal forma, es posible concluir que la regresión logística, en promedio, acierta un 3% más que los árboles de decisión de acuerdo con este planteamiento.

# Exercise 9.5

Ensemble using weighted voting using the oob_error

Evaluate using the following metrics:
* Accuracy
* F1-Score

```python
from sklearn.ensemble import BaggingClassifier
clf = BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=None,random_state=7), n_estimators=100, bootstrap=True,
                        random_state=521, n_jobs=-1, oob_score=True)
clf.fit(X_train, y_train)
y_pred10 = clf.predict(X_test)
print("\033[1;38mDecision Tree max_depth=none")
print("F1: ", metrics.f1_score(y_pred10, y_test), "Accuracy: ",metrics.accuracy_score(y_pred10, y_test))
```

```python
errors = np.zeros(clf.n_estimators)
y_pred_all_ = np.zeros((X_test.shape[0], clf.n_estimators))

for i in range(clf.n_estimators):
    oob_sample = ~clf.estimators_samples_[i]
    y_pred_ = clf.estimators_[i].predict(X_train.values[oob_sample])
    errors[i] = metrics.accuracy_score(y_pred_, y_train.values[oob_sample])
    y_pred_all_[:, i] = clf.estimators_[i].predict(X_test)
    
alpha = (1 - errors) / (1 - errors).sum()
y_pred11 = (np.sum(y_pred_all_ * alpha, axis=1) >= 0.5).astype(np.int)
```

```python
print("\033[1;38mDecision Tree max_depth=none WV")
print("F1: ", metrics.f1_score(y_pred11, y_test), "Accuracy: ",metrics.accuracy_score(y_pred11, y_test))
```

```python
errors = np.zeros(clf2.n_estimators)
y_pred_all_1 = np.zeros((X_test.shape[0], clf2.n_estimators))

for i in range(clf2.n_estimators):
    oob_sample = ~clf2.estimators_samples_[i]
    y_pred_ = clf2.estimators_[i].predict(X_train.values[oob_sample])
    errors[i] = metrics.accuracy_score(y_pred_, y_train.values[oob_sample])
    y_pred_all_1[:, i] = clf2.estimators_[i].predict(X_test)
    
alpha = (1 - errors) / (1 - errors).sum()
y_pred12 = (np.sum(y_pred_all_1 * alpha, axis=1) >= 0.5).astype(np.int)
```

```python
print("\033[1;38mDecision Tree max_depth=2 WV")
print("F1: ", metrics.f1_score(y_pred12, y_test), "Accuracy: ",metrics.accuracy_score(y_pred12, y_test))
```
